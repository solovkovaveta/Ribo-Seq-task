```
data$cor = rep(NA, 16257)
for (i in (1:nrow(data))){
  data$cor[i] = cor.test(as.numeric(data[i,sort(c(seq(3,42,by=4),seq(4,42,by=4)))]), as.numeric(data[i,sort(c(seq(5,42,by=4),seq(6,42,by=4)))]), method = 'spearman')[4]
}
```

```
data$cor
```

Использовала корреляцию Спирмана, а не Пирсона, так как в случае с ненормированными каунтами гораздо более показательными являются данные о порядковых различиях, а не о самом числе каунтов. Впрочем, можно было бы нормировать или логарифмировать данные, и тогда подошла бы корреляция Пирсона.

```
hist(as.numeric(data$cor))
```

Логично предположить, что высокие уровни корреляции характерны для тех генов, у которых транскрипция и трансляция происходят примерно на одинаковом уровне. При этом при низкой корреляции можно предполагать наличие проблем, связанных с, например, деградацией мРНК. 
Такие вещи могут происходить, напримео, при развитии раковой опухоли, посему гены с низкой корреляцией в этом смысле могут являться более интересными.

```
library(edgeR)
library(HTSFilter)
library(limma)
```

```
file = colnames(data[,3:42])
condition = rep(c('normal','tumor'), 20)
samInf = cbind(file, condition)
```

```
rawCount = as.matrix(data[,3:42])
rownames(rawCount) = data[,1]
```

```
rawCountRibo = rawCount[,c(1,2,5,6,9,10,13,14,17,18,21,22,25,26,29,30,33,34,37,38)]
rawCountRNA = rawCount[,c(3,4,7,8,11,12,15,16,19,20,23,24,27,28,31,32,35,36,39,40)]

samInfRibo = samInf[c(1,2,5,6,9,10,13,14,17,18,21,22,25,26,29,30,33,34,37,38),]
samInfRNA = samInf[c(3,4,7,8,11,12,15,16,19,20,23,24,27,28,31,32,35,36,39,40),]
```

```
samInfRNA
```

Сделаем сначала для RNA

```
dgeFull = DGEList(rawCountRNA, group=samInfRNA[,2])
dgeFull
```

```
pseudoCounts = log2(dgeFull$counts+1)
head(pseudoCounts)
plotMDS(pseudoCounts)
```

```
sampleDists =  as.matrix(dist(t(pseudoCounts)))
sampleDists
```

```
dgeFull =  DGEList(dgeFull$counts[apply(dgeFull$counts, 1, sum) != 0, ],
                   group=dgeFull$samples$group)
head(dgeFull$counts)
```

```
dgeFull = calcNormFactors(dgeFull, method="TMM")
dgeFull$samples
```

```
eff.lib.size = dgeFull$samples$lib.size*dgeFull$samples$norm.factors
normCounts = cpm(dgeFull)
pseudoNormCounts = log2(normCounts + 1)
boxplot(pseudoNormCounts, col="gray", las=3)
```

```
dgeFull = estimateCommonDisp(dgeFull)
dgeFull = estimateTagwiseDisp(dgeFull)
dgeFull
```

```
dgeTest = exactTest(dgeFull)
dgeTest
```

```
filtData = HTSFilter(dgeFull)$filteredData
```

```
dgeTestFilt = exactTest(filtData)
dgeTestFilt
```

```
resNoFilt = topTags(dgeTest, n=nrow(dgeTest$table))
head(resNoFilt)
resFilt = topTags(dgeTestFilt, n=nrow(dgeTest$table))
head(resFilt)
```

```
sum(resNoFilt$table$FDR < 0.01)
sum(resFilt$table$FDR < 0.01)
```

```
sigDownRegRNA = resFilt$table[resFilt$table$FDR<0.01,]
sigDownRegRNA = sigDownRegRNA[order(sigDownRegRNA$logFC),]
head(sigDownRegRNA)
```

```
sigUpRegRNA = sigDownRegRNA[order(sigDownRegRNA$logFC, decreasing=TRUE),]
head(sigUpRegRNA)
```

Собственно, в этих двух объектах и хранятся результаты. 

```
plotSmear(dgeTestFilt,
          de.tags = rownames(resFilt$table)[which(resFilt$table$FDR<0.01)])
```

```
volcanoData <- cbind(resFilt$table$logFC, -log10(resFilt$table$FDR))
colnames(volcanoData) <- c("logFC", "negLogPval")
head(volcanoData)
```

```
plot(volcanoData, pch=19)
```

```
plot(as.numeric(a[seq(1,nrow(a),by = 3),2]))
plot(as.numeric(a[seq(2,nrow(a),by = 3),2]))
plot(as.numeric(a[seq(3,nrow(a),by = 3),2]))
```

Вообще это непонятный график, а понятно - сделать бы тест

```
wilcox.test(a[seq(1,nrow(a),by=3),2], a[seq(2,nrow(a),by=3),2])
wilcox.test(a[seq(2,nrow(a),by=3),2], a[seq(3,nrow(a),by=3),2])
wilcox.test(a[seq(1,nrow(a),by=3),2], a[seq(3,nrow(a),by=3),2])
```
