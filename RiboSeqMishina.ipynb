Начнем с задания 1 и чтения файлов. Для того, чтобы они таки прочитались, при работе в колабе нужно загрузить данные - я не сохраняла их на свой гугл диск, а загружала непосредственно с устройства.

```
data = read.table('01. RiboSeq_RNASeq_HCC_counts.tsv', header = T)
```

```
data
```

```
data$notna = rep(0, 18945)
data$notna = apply(data, 1, function(x) all(is.na(x[3:42])==F))
```

```
table(data$notna)
```

```
data = data[data$notna == T, ]
```

```
data
```

```
data$cor = rep(NA, 16257)
for (i in (1:nrow(data))){
  data$cor[i] = cor(data[i,c(seq(3,42,by=4),seq(4,42,by=4))], data[i,c(seq(5,42,by=4),seq(6,42,by=4))], method = 'spearman')
}
```

Использовала корреляцию Спирмана, а не Пирсона, так как в случае с ненормированными каунтами гораздо более показательными являются данные о порядковых различиях, а не о самом числе каунтов. Впрочем, можно было бы нормировать или логарифмировать данные, и тогда подошла бы корреляция Пирсона.
